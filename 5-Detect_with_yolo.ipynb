# بارگذاری کتابخانه ها

```python
import cv2,time
from ultralytics import YOLO

# مسیر فایل مدل از پیش دانلود شده
model_path = 'YOLO_V8_models/yolov8n.pt'

# بارگذاری مدل از مسیر فایل
model = YOLO(model_path)
```

# باز کردن وب کم و سوار کردن مدل از پیش آموزش دیده

```python
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)

    # YOLOv8 استفاده از مدل برای پردازش هر فریم
    results = model(frame)

    # رسم نتایج روی فریم
    annotated_frame = results[0].plot()  # نقاشی کردن نتایج
    
    cv2.imshow('Webcam', annotated_frame)


    k = cv2.waitKey(1)
    if k == ord('q'):
        break
    elif k == 27:
        break
    elif k == ord('s'):
        cv2.imwrite(f'screenshot{time.time()}.png', annotated_frame)
        print('screenshot saved')
        
cap.release()
cv2.destroyAllWindows()
```
