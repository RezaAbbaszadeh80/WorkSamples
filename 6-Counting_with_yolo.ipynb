```python
# استفاده از مدل دانلود شده
from ultralytics import YOLO

# مسیر فایل مدل از پیش دانلود شده
model_path = 'YOLO_V8_models/yolov8n.pt'

# بارگذاری مدل از مسیر فایل
model = YOLO(model_path)
```

```python
import matplotlib.pyplot as plt
import cv2 

# بارگذاری و تشخیص اشیا در تصویر
results = model('Data/car_plate_russia.jpg')

# شمارش تعداد اشیا
objects_detected = len(results[0].boxes)
print(f"Number of objects detected: {objects_detected}")

# استخراج تصویر با جعبه‌های اطراف اشیاء شناسایی شده
annotated_image = results[0].plot()

# نمایش داده میشه ولی توی وبکم لازم نبود matplotlib باز شده داره با opencv چون چیزی که با  RGB به BGR
annotated_image = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

# matplotlib نمایش تصویر با استفاده از 
plt.figure(figsize=(12, 8))  # تنظیم ابعاد به 12 اینچ در 8 اینچ
plt.imshow(annotated_image)
plt.axis('off')  # حذف محورها
plt.show()
```
