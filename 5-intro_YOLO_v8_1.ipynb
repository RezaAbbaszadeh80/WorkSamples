# کتابخانه‌های مرتبط با وب‌کم و بینایی ماشین در پایتون

### 1. OpenCV
- کتابخانه محبوب برای پردازش تصویر و ویدیو، شامل تشخیص اشیا و چهره
- نصب:
  ```bash
  pip install opencv-python
  ```

### 2. Mediapipe
- ابزار قدرتمند برای ردیابی و تشخیص اجزای بدن، ژست‌ها و حالت‌های صورت
- نصب:
  ```bash
  pip install mediapipe
  ```

### 3. dlib
- کتابخانه‌ای برای تشخیص چهره و نقاط کلیدی صورت
- نصب:
  ```bash
  pip install dlib
  ```

### 4. PyTorch و TensorFlow
- فریم‌ورک‌های یادگیری عمیق برای ساخت و آموزش مدل‌های بینایی ماشین
- نصب:
  ```bash
  pip install torch torchvision
  pip install tensorflow
  ```

### 5. YOLO (You Only Look Once)
- مدل تشخیص اشیا در تصاویر و ویدیوها با قابلیت بلادرنگ
- نصب برای YOLOv8:
  ```bash
  pip install ultralytics
  ```

### 6. Scikit-image
- کتابخانه پردازش تصویر برای پروژه‌های بینایی ماشین ساده و متوسط
- نصب:
  ```bash
  pip install scikit-image
  ```

### 7. Face Recognition
- کتابخانه ساده برای تشخیص چهره‌ها در تصاویر و ویدیوها
- نصب:
  ```bash
  pip install face_recognition
  ```

### 8. SimpleCV
- کتابخانه سطح بالا برای پردازش تصویر و کار با وب‌کم
- نصب:
  ```bash
  pip install SimpleCV
  ```
