# Types of mode (Mode) in YOLOv8

### Predict: پیش‌بینی و گرفتن خروجی از مدل آموزش‌دیده
### Train: آموزش مدل
### Val: ارزیابی مدل و نمایش معیارهای دقت و غیره
### Export: ذخیره مدل مثلا برای توسعه
### Benchmark: اجرای ارزیابی‌های مختلف و رایج مانند سرعت، دقت، mAP و غیره
### Track: ردیابی اشیا در ویدئوه
